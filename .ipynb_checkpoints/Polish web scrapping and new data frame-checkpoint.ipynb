```python
import pandas as pd
import numpy as np
```

```python
old_stats = pd.read_csv('scrapDriversStatsAll.csv')
```

```python
old_stats.head()
```

```python
#Create a new dataframe with the columns that we need.
```

```python
new_stats = old_stats[['Nombre','Championships','Grands','Grand','wins','win','pole','fastests','fastest','podiums','podium','retirements','retirement','constructors','constructor','models','model']].copy()
```

```python
new_stats.head()
```

```python
#Save the new dateframe.
```

```python
new_stats.to_csv('driversDataFrame2.csv')
```

```python
#Replace NaN with 0
```

```python
new_stats.fillna(0)
```

```python
new_stats.head()
```

```python
#Select columns with float type to covert to int, then to string.
```

```python
col =['Grands','wins','win','pole','fastests','fastest','podiums','podium','retirements','retirement','constructors','constructor','models','model']
```

```python
#converting to int. Fill NaN with zeros and import numpy first to aplly int64 mapping.
```

```python
new_stats[col] = new_stats[col].fillna(0).applymap(np.int64)
```

```python
new_stats
```

```python
#Save the new dateframe again.
```

```python
new_stats.to_csv('driversDataFrame2.csv')
```

```python
#Convert columns to string.
```

```python
for i in col:
    new_stats[i] = new_stats[i].astype(str)
```

```python
#Save the new dateframe again.
```

```python
new_stats.to_csv('driversDataFrame.csv')
```

```python
#Join some columns with similar names. Keep those written in plural.
```

```python
new_stats["wins"] = new_stats["wins"] + new_stats["win"]
```
